# Ejercicio I - Periodograma

```python
%matplotlib notebook

from numpy.fft import *
import numpy as np
import matplotlib.pyplot as plt

#definimos constantes
N = 32
Nfft = 1024
```

```python
#definir funcion

def periodograma(x,Nfft):
    """
    Function that computes periodogram
    Input
     - x:signal
     - nfft: number of points in the fft
    Output
     - psd: power spectral density
     """
    
    N = len(x)
    
    x_f = np.fft.fft(x,Nfft)

    Px = np.abs(x_f)**2/N    # Periodograma
    
    Px = np.fft.fftshift(Px)
    
    f = np.fft.fftshift(np.fft.fftfreq(Nfft))
    
    return Px,f
```

```python
np.random.seed(1947)  # Dejamos fija la ejecuciÃ³n. Aunque ejecutemos, tendremos la misma grÃ¡fica.

noise = np.random.randn(N)
Px, f = periodograma(noise, Nfft)

idx = f >= 0          # Mostramos las frecuencias positivas
plt.plot(f[idx], 10*np.log10(Px[idx]))

plt.xlabel('Freq [u.n.]')
plt.ylabel('P [dB]')
```
